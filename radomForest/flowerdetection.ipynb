```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split as tts
from sklearn.ensemble import RandomForestClassifier
import seaborn as sns
```

```python
flower = load_iris()
```

```python
dir(flower)
```

```python
df = pd.DataFrame(flower.data, columns = flower.feature_names)
```

```python
df['target'] = flower.target
```

```python
df['flower name'] = df.target.apply(lambda x: flower.target_names[x])
```

```python
y = df.target
```

```python
df.drop(['target', 'flower name'], axis = 1, inplace = True)
```

```python
X=df
```

```python
model = RandomForestClassifier(n_estimators = 10)
```

```python
X_train, X_test, y_train, y_test = tts(X, y, test_size = 0.2)
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot = True, cmap = 'hot', xticklabels = flower.target_names, yticklabels =  flower.target_names)
```
