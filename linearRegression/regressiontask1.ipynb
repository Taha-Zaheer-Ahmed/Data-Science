```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head(3)
```

```python
plt.scatter(x = df['per capita income (US$)'], y = df['year'], color = 'red', marker = '.')
plt.xlabel('Per Capita income (US$)', fontsize = 20)
plt.title('Canada Per Capita income over the years', fontsize = 20)
plt.ylabel('Year', fontsize = 20)
plt.show()
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
np.arange(start = 2017, stop = 2021)
```

```python
future_df = pd.DataFrame(np.arange(start = 2017, stop = 2031), columns = ['year'])
```

```python
plt.scatter(x = df['per capita income (US$)'], y =  df['year'], color = 'red', marker = '.', label = 'Data')
plt.xlabel('Per Capita income (US$)', fontsize = 20)
plt.title('Canada Per Capita income (1970 - 2016)', fontsize = 20)
plt.ylabel('Year', fontsize = 20)
plt.plot(reg.predict(df[['year']]), df['year'], color = 'b', label = 'Linear Regression Fit')
plt.legend()
plt.show()
```

```python
future_df['per capita income (US$)'] = reg.predict(future_df[['year']])
```

```python
future_df
```

```python
new_df = pd.merge(df, future_df, on = ['year', 'per capita income (US$)'], how = 'outer' )
```

```python
plt.plot(new_df['per capita income (US$)'], new_df['year'], color = 'red', marker = '.')
plt.xlabel('Per Capita income (US$)', fontsize = 20)
plt.title('Canada Per Capita income over the years (1970 - 2030)', fontsize = 20)
plt.ylabel('Year', fontsize = 20)
```
