```python
def people_converter_cell(cell):
    if cell == 'n.a.':
        return 'sam walton'
    return cell
    

df2 = pd.read_excel('stock_data.xlsx', converters = {'people': people_converter_cell}, na_values = {'eps': ['not available', 'n.a.'], 'revenue' :  ['not available', 'n.a.', -1], 'price':  ['not available', 'n.a.', -1], 'people':  ['not available', 'n.a.']})
df2
```
