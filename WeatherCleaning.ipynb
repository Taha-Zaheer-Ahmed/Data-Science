```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#fillna() method, fills NaN values with specified elements 
#Value argument is the value to be filled in Nan cells
#inplace argument resets the original dataframe
df1 = pd.read_csv('NYC_weather_data.csv')
plt.title('DataFrame data')
plt.imshow(df1[['temperature', 'windspeed']], cmap = "autumn")
plt.show()
s = df1.fillna(value = 0)
s
```

```python
#Values can also be given in the form of dictionary as shown below
df1 = pd.read_csv('NYC_weather_data.csv')
value = {'temperature': 0, 'windspeed':0, 'event': 'undetermined'}
df1.fillna(value = value, inplace = True)
df1
plt.title('DataFrame data')
plt.imshow(df1[['temperature', 'windspeed']], cmap = "autumn")
plt.show()
```

```python
df1.sort_values(by = 'temperature', ascending = False, inplace = True)
plt.title('DataFrame data')
plt.imshow(df1[['temperature', 'windspeed']], cmap = "autumn")
plt.show()
df1
```

```python
#Since the dates obtained from CSV files are considered Strings, therefore we use parse_dates 
#argument to convert the columns of 'day' in a data type of 'Dates'
df1 = pd.read_csv('NYC_weather_data.csv', parse_dates = ['day'])
df1.set_index('day', inplace = True)
df1
```

```python
#the ffill method carry forwards the previous value in the dataframe
#column for NaN values
#limit argument limits the amount of times a method can copy a certain value
df1 = pd.read_csv('NYC_weather_data.csv', parse_dates = ['day'])
df1.set_index('day', inplace = True)
s = df1.fillna(method = 'ffill')
plt.title('DataFrame Data')
plt.imshow(s[['temperature', 'windspeed']], cmap = "autumn" )
plt.show()
s
```

```python
df1 = pd.read_csv('NYC_weather_data.csv', parse_dates = ['day'])
df1.set_index('day', inplace = True)
df1.interpolate(method = 'time')
```

```python
#dropna() function drops all NaN values 
#the how argument provides the limit for how many rows to drop
#based on the number of NaNs
df1 = pd.read_csv('NYC_weather_data.csv', parse_dates = ['day'])
df1.set_index('day', inplace = True)
df1.dropna(how = "all")
```

```python
#the threshold argument provides the minumum number of a Non-NaN values in 
#a row to not be dropped 
df1 = pd.read_csv('NYC_weather_data.csv', parse_dates = ['day'])
df1.set_index('day', inplace = True)
df1.dropna(thresh = 2)
```

```python
#In order to add missing date values we can make a data range and
#reindex our dataframe
df1 = pd.read_csv('NYC_weather_data.csv', parse_dates = ['day'])
df1.set_index('day', inplace = True)
dt = pd.date_range('2017-01-01', '2017-01-11')
idx = pd.DatetimeIndex(dt)
df1 = df1.reindex(idx)
```
