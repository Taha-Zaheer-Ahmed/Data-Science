```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LogisticRegression
import seaborn as sns
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
plt.imshow(digits.data[94].reshape(8,8), cmap = 'hot')
```

```python
X_train, X_test, y_train, y_test = tts(digits.data, digits.target, test_size = 0.2)
```

```python
model1 = LogisticRegression(solver='lbfgs', max_iter=1000)
```

```python
model1.fit(X_train, y_train)
```

```python
model1.score(X_test, y_test) * 100
```

```python
y_pred = model1.predict(X_test)
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot = True, cmap= 'hot')
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.title('Confusion Matrix of Image Dataset')
plt.show()
```
