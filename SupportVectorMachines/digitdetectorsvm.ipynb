```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split as tts
from sklearn.svm import SVC
import seaborn as sns
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits.data.shape
```

```python
sns.heatmap(digits.images[179], cmap= 'Blues')
plt.show()
sns.heatmap(digits.images[176], cmap= 'Blues')
plt.show()
sns.heatmap(digits.images[189], cmap= 'Blues')
plt.show()
sns.heatmap(digits.images[149], cmap= 'Blues')
plt.show()
```

```python
X = digits.data
```

```python
y = digits.target
```

```python
X_train, X_test, y_train, y_test = tts(X, y, test_size = 0.2)
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict([digits.data[174]])
```

```python
y_pred = model.predict(X_test)
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred), annot = True, cmap = 'mako')
plt.xlabel('True Values')
plt.ylabel('Predicted Values')
plt.title('Confusion Matrix')
```
