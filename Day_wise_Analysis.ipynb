```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
import numpy as np
```

```python
df = pd.read_csv("day_wise.csv")
df
```

```python
df.columns
```

```python
df.drop(['Deaths / 100 Cases','Recovered / 100 Cases','Deaths / 100 Recovered'], inplace = True, axis = 1)
df
```

```python
df
```

```python
df.set_index(df.Date, inplace= True, drop = True)
df
```

```python
df.shape
```

```python
plt.plot(df.Deaths, color = 'r', label = "Deaths")
plt.plot(df.Recovered, color = 'b', label = 'Recovered')
plt.plot(df.Active, color = 'y', label = 'Active')
plt.xticks(rotation=45, ha='right')
plt.xticks(np.arange(0, df.shape[0] + 1, (df.shape[0] + 1)//11))
plt.gca().yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
plt.gca().ticklabel_format(axis='y', style='plain')
plt.xlabel("Date", fontsize = 12)
plt.ylabel("Number of People", fontsize = 12)
plt.title("Covid-19 Analysis", fontsize = 24)
plt.legend()
plt.show()
```
